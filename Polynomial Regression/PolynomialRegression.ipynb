```python
# If the data-points clearly will not fit a linear regression (a straight line through all data points), 
# it might be ideal for polynomial regression.
```

```python
import matplotlib.pyplot as plt

# time of day (hour)
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
# car speed
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

plt.scatter(x, y)
plt.show()    
```

```python
# drawing polynomial regression line using numpy method
import numpy

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

myline = numpy.linespace(1, 22, 100)

plt.scatter(x, y)

plt.plot(myline, mymodel(myline))

plt.show()
```
